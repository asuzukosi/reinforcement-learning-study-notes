```python
# testing the use of gymnasium within a notebook
import gymnasium as gym
```

```python
len(gym.envs.registry.keys()) # returns the number of possible environment markov decision processes
```

```python

# At the high level is the environment which 
# represents the markov decision process
env = gym.make('LunarLander-v2', render_mode='human')
observation, info = env.reset()


for _ in range(1000):
    action = env.action_space.sample()
    print(action)
    observation, reward, trauncated, terminated, info = env.step(action)
    if terminated or trauncated:
        observation, info  = env.reset()
        
env.close()
```
